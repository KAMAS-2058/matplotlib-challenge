```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)



# Display the data table for preview
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata,study_results,on ="Mouse ID")
```

```python
# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
len(combined_data['Mouse ID'].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
combined_data.duplicated(subset=['Mouse ID', 'Timepoint'])
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#took the last measurment to see the full effects of the drug Regimen
cleaned_mice = combined_data.drop_duplicates(subset=['Mouse ID','Timepoint'])
cleaned_mice
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_mice['Mouse ID'].unique())
```

## Summary Statistics

```python
cleaned_mice
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean_tu = cleaned_mice.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()
med_tu = cleaned_mice.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()
var_tu = cleaned_mice.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()
std_tu = cleaned_mice.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
sem_tu = cleaned_mice.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()

mice_effects = pd.DataFrame({
    'Mean Tumor by Drug, (mm3)' : mean_tu,
   ' Median Tumor by Drug, (mm3)': med_tu,
    'Tumor Variance by Drug, (mm3)': var_tu,
    'Standard Dev by Drug, (mm3)': std_tu,
    'SEM by Drug, (mm3)' : sem_tu,   
})
mice_effects
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
clean_mice_bar = cleaned_mice.groupby(['Drug Regimen'])['Timepoint'].sum()
clean_mice_bar.plot.bar(y=['Timepoint'])
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
yaxis = clean_mice_bar
fig, bone = plt.subplots()
bone.bar(clean_mice_bar.index, yaxis)
bone.set_xticklabels(mice_effects.index ,rotation = 45, horizontalalignment = 'right')
bone.set_title('Number of Timepoints measure based on Drug Regimen', Fontsize = 22)
plt.xlabel('Drug Administered')
plt.ylabel('Number of Timepoints by Drug')
plt.show
```

## Quartiles, Outliers and Boxplots

```python
# # Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# # Capomulin, Ramicane, Infubinol, and Ceftamin

the_four = cleaned_mice.loc[(cleaned_mice['Drug Regimen'] == 'Capomulin') | (cleaned_mice['Drug Regimen'] == 'Ramicane') |
                                            (cleaned_mice['Drug Regimen'] == 'Infubinol') |(cleaned_mice['Drug Regimen'] == 'Ceftamin') ]
# # Start by getting the last (greatest) timepoint for each mouse
the_four =the_four.groupby(['Drug Regimen','Mouse ID']).max()['Timepoint']

# # Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_four = pd.merge(the_four, cleaned_mice, on= ('Mouse ID','Timepoint'), how = 'left')
    
merged_four
```
